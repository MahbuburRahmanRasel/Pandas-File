```python
import pandas as pd 
ufo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/ufo.csv")
```

```python
ufo.head()
```

```python
ufo.loc[0,:]
```

```python
ufo.loc[[0,2,4],:]
```

```python
ufo.loc[0:2,:] # include both side of notation
```

```python
ufo.loc[:,"City"]
```

```python
ufo.loc[:,['City','State']]
```

```python
ufo.loc[:,'City':'State']
```

```python
ufo.loc[0:2,'City':'State']
```

```python
ufo[ufo.City=='Oakland']
```

```python
ufo.loc[ufo.City=='Oakland',:].head()
```

```python
ufo.loc[ufo.City=='Oakland','State']
```
